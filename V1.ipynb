```python
# List of binary symptom columns
symptom_cols = ['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']

# Function to get present symptoms for each row
def extract_symptoms(row):
    return [symptom for symptom in symptom_cols if row[symptom] == 'Yes']

df['Symptoms_List'] = df.apply(extract_symptoms, axis=1)
```

```python
def make_document(row):
    symptoms_text = ', '.join(row['Symptoms_List']) or "no major symptoms"
    return (
        f"Disease: {row['Disease']}.\n"
        f"Symptoms: {symptoms_text}.\n"
        f"Patient profile: Age {row['Age']}, {row['Gender'].lower()}, "
        f"Blood Pressure: {row['Blood Pressure']}, Cholesterol: {row['Cholesterol Level']}."
    )

df['Document'] = df.apply(make_document, axis=1)

# Show example
df[['Document']].head()
```
